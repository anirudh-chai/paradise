```python
from face2face import Face2Face
f2f = Face2Face(device_id=0)
```

```python
swapped_img = f2f.swap_img_to_img(r"C:\Users\91988\Downloads\clearsrk.jpg", r"C:\Users\91988\Downloads\paradise nani.jpg")
```

```python
import cv2
import matplotlib.pyplot as plt

# Convert BGR to RGB
swapped_img_rgb = cv2.cvtColor(swapped_img, cv2.COLOR_BGR2RGB)

# Display the corrected image
plt.figure(figsize=(10, 8))
plt.imshow(swapped_img_rgb)
plt.axis('off')
plt.show()
```

```python
# create a face embedding and save it to disk. 
embedding = f2f.add_face("my_new_face", r"C:\Users\91988\Downloads\prabhas.jpg", save=True)
# Swap all faces in the image or video with the face(s) in the face embedding
swapped = f2f.swap(media=r"C:\Users\91988\Downloads\subtle.mp4", faces="my_new_face")
```

```python
output_path = r"C:\Users\91988\Downloads\face_swapped_video2.mp4"
swapped.save(output_path)
print(f"Video saved to: {output_path}")
```
